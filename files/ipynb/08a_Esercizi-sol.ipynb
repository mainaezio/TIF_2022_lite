__1 - Fate il plot di $\frac{\sin (x)}{x}$. Scegliete un intervallo opportuno. Cosa succede nell'intorno di $x \,=\, 0$?__

__2 - Fate il plot di $\tan (x)$.__ (D.J.Pine, Introduction to Python for Science and Engineering, 6.2.3)

```python
import matplotlib.pyplot as plt
import numpy as np
```

Primo tentativo:

```python
xc = np.arange(-4.0, 4.0, 1/10.)
#xc = np.arange(-4.0, 4.0, 1/100.)
yc =np.tan(xc)

fig, ax = plt.subplots(figsize=(8, 5))
ax.set_xlabel('x')
ax.set_ylabel('tan(x)')
ax.plot(xc, yc, c='k')
```

Secondo tentativo:

```python
xc = np.arange(-4.0, 4.0, 1/1000.)
xc = [x for x in xc if np.abs(np.tan(x)) < 10.]

yc =np.tan(xc)

fig, ax = plt.subplots(figsize=(8, 5))
ax.set_xlabel('x')
ax.set_ylabel('tan(x)')
ax.plot(xc, yc, c='k')
```

__3 - Fare il grafico della funzione $\int_0^x \, \frac{\sin(t)}{t} \,dt$
fra $x = -10$ e $x = 10$.__

```python
import sympy
x,y = sympy.symbols('x,y')
sympy.integrate(sympy.sin(y)/y,(y,x,0))
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x_arr = np.arange(-10,10,0.1)
y_arr = np.array([-sympy.Si(t) for t in x_arr])
ax.plot(x_arr,y_arr)
```

__6 - Disegnate,in un solo grafico, la funzione $f(x) =  \frac{x\,(x-1)}{x**2+1}-1$, la sua derivata prima e la sua derivata seconda nell'intervallo (-8,8).__

```python
from sympy import *
init_printing()
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = symbols('x')
f = x*(x-1)/(x**2+1)-1
f1 = diff(f,x)
f2 = diff(f1,x)
```

```python
f
```

```python
xc = np.arange(-8.0, 8.0, 1/10.)
f_lam = lambdify(x,f)
yc = f_lam(xc)
```

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.margins(x=0.,y=0.1)
ax.grid(True)
ax.plot(xc, yc)
```

```python
f1_lam = lambdify(x,f1)
yc1 = f1_lam(xc)
```

```python
f2_lam = lambdify(x,f2)
yc2 = f2_lam(xc)
```

Zeri di f':

```python
solve(f1,x)
```

Zeri di f'':

```python
solve(f2,x)
```

```python
fig, ax = plt.subplots(figsize=(14, 8))
ax.set_xlabel('x')
ax.set_ylabel('f(x),f\'(x),f\"(x)')
ax.margins(x=0.,y=0.1)
ax.grid(True)
ax.plot(xc, yc, c='b', label='f')
ax.plot(xc, yc1, c='g', label="f'")
ax.plot(xc, yc2, c='r', label='f"')
ax.legend(fontsize='20')
ax.margins(x=0.,y=0.05)
ax.vlines(-1-np.sqrt(2.),-1.5,2.5, colors='g',linestyles='dashed') # zero di f'
ax.vlines(-1+np.sqrt(2.),-1.5,2.5, colors='g',linestyles='dashed') # zero di f' 
ax.vlines(1,-1.5,2.5, colors='r', linestyles='dotted')             # zero di f''
ax.vlines(-2-sqrt(3.),-1.5,2.5, colors='r',linestyles='dotted')    # zero di f''
ax.vlines(-2+sqrt(3.),-1.5,2.5, colors='r',linestyles='dotted')    # zero di f''
```
